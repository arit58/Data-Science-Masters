Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

```python
from scipy import stats
sample_mean = 50
sample_std = 5
confidence_level = 0.95
sample_size = 100

std_error = sample_std/(sample_size**0.5)

lower_limit, upper_limit = stats.norm.interval(confidence_level,loc=sample_mean,scale=std_error)

print(f"95% confidence interval: ({lower_limit:.2f},{upper_limit:.2f})")
```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the
population who smoke.

```python
import math

# sample size
n = 500

# number of individuals who smoke in the sample
x = 60

# sample proportion
p_hat = x / n

# z-score for 95% confidence level
z = 1.96

# standard error of the proportion
SE = math.sqrt(p_hat * (1 - p_hat) / n)

# margin of error
ME = z * SE

# lower and upper bounds of the confidence interval
lower_bound = (p_hat - ME)*100
upper_bound = (p_hat + ME)*100

# print the confidence interval
print("The 95% confidence interval is: ({:.4f}, {:.4f})".format(lower_bound, upper_bound))
```

Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation
of 12 using Python. Interpret the results.

```python
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 30
alpha = 0.1

t_value = stats.t.ppf(1 - alpha/2, df=sample_size-1)
margin_of_error = t_value * sample_std / (sample_size ** 0.5)
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("The 90% confidence interval is:", confidence_interval)
```

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10.
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a
significance level of 0.05.

To conduct a hypothesis test for determining if the population mean is significantly different from 70, we need to follow the steps below:

The null hypothesis (H0) states that the population mean is equal to 70. The alternative hypothesis (Ha) states that the population mean is significantly different from 70.

H0: μ = 70
Ha: μ ≠ 70

The level of significance (α) is given as 0.05.

We can use the t-test since the population standard deviation is unknown. The formula for the t-test statistic is:

t = (x̄ - μ) / (s / √n)

where x̄ is the sample mean, μ is the population mean, s is the sample standard deviation, and n is the sample size.

Substituting the given values, we get:

t = (72 - 70) / (10 / √30) = 1.095

Since the alternative hypothesis is two-tailed, we need to find the critical value for a two-tailed test with α = 0.05 and degrees of freedom (df) = n - 1 = 29.

Using a t-distribution table or a calculator, we find the critical values as ±2.045.

Since the calculated t-value (1.095) is less than the critical value (±2.045), we fail to reject the null hypothesis.

Based on the sample data, we do not have sufficient evidence to conclude that the population mean is significantly different from 70 at the 0.05 level of significance.
