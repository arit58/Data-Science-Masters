Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test. <br>
A t-test and a z-test are both statistical tests used to make inferences about population parameters based on sample data. However, they differ in a few key ways:

Population Variance: A z-test assumes that the population variance is known, while a t-test assumes that the population variance is unknown.

Sample Size: A z-test is appropriate when the sample size is large (typically greater than 30), while a t-test is appropriate for smaller sample sizes.

Normal Distribution: A z-test assumes that the population is normally distributed, while a t-test assumes that the population is approximately normally distributed.

Hypothesis Testing: In a z-test, the null hypothesis is usually that the population mean is equal to a specified value, while in a t-test, the null hypothesis is usually that the population mean is equal to a specified value or that two population means are equal.

Test Statistic: The test statistic for a z-test is a z-score, while the test statistic for a t-test is a t-score.<br>

z test is used when the population standard deviation is known, and the sample size is more than 30.
t test is used when the population standard deviation is unknown or the sample size is small.

Q2: Differentiate between one-tailed and two-tailed tests.<br>
In a one-tailed test, the alternative hypothesis specifies the direction of the effect, either an increase or a decrease in the population parameter. For example, a one-tailed test may test whether a new drug increases the average lifespan of a patient. The null hypothesis in this case would be that the drug has no effect on lifespan, while the alternative hypothesis would be that the drug increases lifespan. The one-tailed test will reject the null hypothesis if the test statistic falls in the critical region in the direction of the alternative hypothesis.

In a two-tailed test, the alternative hypothesis specifies that the effect can occur in either direction. For example, a two-tailed test may test whether a coin is fair, where the null hypothesis is that the coin is fair and the alternative hypothesis is that the coin is biased, either towards heads or tails. The two-tailed test will reject the null hypothesis if the test statistic falls in either tail of the distribution.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error. <br>
Type 1 error occurs when the null hypothesis is rejected even though it is actually true. In other words, it is a false positive. This error occurs when we set the level of significance (α) too low, and we reject the null hypothesis when it is actually true. The probability of making a Type 1 error is denoted by α and is also called the level of significance. The level of significance is typically set at 0.05 or 0.01.

Example scenario: A pharmaceutical company is testing a new drug to treat a certain disease. The null hypothesis is that the drug has no effect on the disease. The company conducts a clinical trial and concludes that the drug is effective in treating the disease when in reality, it is not. This is a Type 1 error.<br>
Type 2 error occurs when the null hypothesis is not rejected even though it is actually false. In other words, it is a false negative. This error occurs when we set the level of significance too high, and we fail to reject the null hypothesis when it is actually false. The probability of making a Type 2 error is denoted by β.

Example scenario: A medical researcher is studying the relationship between smoking and lung cancer. The null hypothesis is that there is no relationship between smoking and lung cancer. The researcher conducts a study and fails to find a significant relationship between smoking and lung cancer when in reality, there is a significant relationship. This is a Type 2 error.

Q4: Explain Bayes's theorem with an example. <br>
Bayes's theorem is a mathematical formula used to update the probability of a hypothesis based on new evidence or information. The theorem is named after the English mathematician Thomas Bayes, who developed it in the 18th century.

The formula for Bayes's theorem is:

P(A|B) = P(B|A) x P(A) / P(B)

where:

P(A|B) is the probability of hypothesis A given evidence B
P(B|A) is the probability of evidence B given hypothesis A
P(A) is the prior probability of hypothesis A (i.e., the probability of A before considering evidence B)
P(B) is the prior probability of evidence B (i.e., the probability of B before considering hypothesis A)
To illustrate Bayes's theorem, consider the following example:

Suppose a company produces two types of light bulbs: Type A and Type B. 80% of the light bulbs produced are Type A, and 20% are Type B. Type A bulbs have a failure rate of 5%, while Type B bulbs have a failure rate of 10%.

Now, suppose you randomly select a bulb from the company's inventory and it fails. What is the probability that the bulb is Type A?

Using Bayes's theorem, we can calculate the probability as follows:

P(A) = 0.8 (prior probability of selecting a Type A bulb)
P(B) = 0.2 (prior probability of selecting a Type B bulb)
P(failure|A) = 0.05 (probability of a Type A bulb failing)
P(failure|B) = 0.1 (probability of a Type B bulb failing)
We want to calculate P(A|failure), the probability that the bulb is Type A given that it failed.

Using Bayes's theorem, we have:

P(A|failure) = P(failure|A) x P(A) / P(failure)

P(failure) can be calculated using the law of total probability:

P(failure) = P(failure|A) x P(A) + P(failure|B) x P(B)
= 0.05 x 0.8 + 0.1 x 0.2
= 0.06

Plugging in the values, we get:

P(A|failure) = P(failure|A) x P(A) / P(failure)
= 0.05 x 0.8 / 0.06
= 0.67

Therefore, the probability that the bulb is Type A given that it failed is 0.67, or 67%.


Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example. <br>
A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain level of confidence. It is a statistical measure of the precision of an estimate or measurement.

For example, suppose a researcher wants to estimate the average height of adult males in a certain population. The researcher takes a random sample of 100 men from that population and calculates the sample mean height to be 180 cm. However, the researcher is aware that the sample mean height is unlikely to be exactly equal to the true population mean height. Therefore, the researcher constructs a 95% confidence interval to provide a range of values in which the true population mean height is likely to fall.

To calculate the confidence interval, the following formula is used:

Confidence interval = sample statistic ± (critical value x standard error)

where:

Sample statistic: The sample mean or proportion being used to estimate the population parameter
Critical value: The value obtained from the standard normal distribution or t-distribution corresponding to the desired level of confidence and degrees of freedom (df)
Standard error: The standard deviation of the sampling distribution, which is calculated as the standard deviation of the sample divided by the square root of the sample size
For example, suppose the researcher wants to construct a 95% confidence interval for the population mean height based on the sample of 100 men with a sample mean height of 180 cm and a sample standard deviation of 5 cm. Using a t-distribution with 99 degrees of freedom (df = n - 1 = 100 - 1 = 99), the critical value for a 95% confidence level is 1.984. The standard error can be calculated as:

Standard error = sample standard deviation / square root of sample size
= 5 / sqrt(100)
= 0.5

Plugging in the values, the confidence interval can be calculated as:

Confidence interval = 180 ± (1.984 x 0.5)
= (179.01, 180.99)

Therefore, the researcher can be 95% confident that the true population mean height falls between 179.01 cm and 180.99 cm.

What is the probability that the product is Type A?

Solution: Let A be the event that the product is Type A, and D be the event that the product is defective. We want to find the probability of A given D, P(A|D).

Using Bayes' Theorem, we have:

P(A|D) = P(D|A) x P(A) / P(D)

where:

P(D|A) is the probability of the product being defective given that it is Type A
P(A) is the prior probability of the product being Type A
P(D) is the probability of the product being defective
We can calculate these probabilities as follows:

P(D|A) = 0.04 (given)
P(A) = 0.60 (given)
P(D) = P(D|A) x P(A) + P(D|B) x P(B) = 0.04 x 0.60 + 0.10 x 0.40 = 0.064
Plugging in the values, we get:

P(A|D) = P(D|A) x P(A) / P(D)
= 0.04 x 0.60 / 0.064
= 0.375

Therefore, the probability that the product is Type A given that it is defective is 0.375 or 37.5%.

Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average
of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is
significantly effective at a 95% confidence level using a t-test. <br>

To conduct a hypothesis test, we need to state the null and alternative hypotheses. In this case, our null hypothesis is that the new weight loss drug is not significantly effective, meaning that the population mean weight loss is equal to 0. Our alternative hypothesis is that the drug is significantly effective, meaning that the population mean weight loss is greater than 0.

Null hypothesis: μ = 0
Alternative hypothesis: μ > 0

We will use a t-test since the sample size is less than 30 and the population standard deviation is unknown.

We will use a significance level (α) of 0.05, meaning that we are 95% confident in our conclusion.

The formula for the t-statistic is:

t = (x̄ - μ) / (s / √n)

where x̄ is the sample mean, μ is the population mean, s is the sample standard deviation, and n is the sample size.

Plugging in the numbers we get:

t = (6 - 0) / (2.5 / √50) = 10.392

Using a t-table with 49 degrees of freedom (df = n - 1), we find that the critical t-value for a one-tailed test with a significance level of 0.05 is 1.677.

Since our calculated t-value of 10.392 is greater than the critical t-value of 1.677, we can reject the null hypothesis and conclude that the new weight loss drug is significantly effective at a 95% confidence level.

Therefore, we can conclude that the drug has a significant effect on weight loss based on the results of this study.

Confidence interval = sample proportion ± margin of error

Margin of error = 1.96 * √((0.65 * 0.35) / 500) ≈ 0.040
