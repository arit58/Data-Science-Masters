Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
<br>
The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

```python
#The IndexError occurs in Python when an item from a list is attempted to be accessed that is outside the index range of the list.
x = [1, 2, 3, 4]
print(x[10])
```

```python
#When working with dictionaries in Python, a KeyError gets raised when you try to access an item that doesn't exist in a Python dictionary.
student = {
  "name": "John",
  "course": "Python",
}
print(student["age"])
```
