Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
import pandas as pd
names = pd.Series(['Messi','Haaland','Pedri'],name='name')
countries = pd.Series(['Arg','Nor','Spain'],name='country')
teams = pd.Series(['psg','mcity','Barca'],name='team')

index_labels=['r1','r2','r3']
names.index = index_labels
countries.index = index_labels
teams.index = index_labels

df = pd.concat([names,countries,teams],axis=1)                
df
```
