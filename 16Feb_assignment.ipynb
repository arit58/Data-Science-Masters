Q1. What is a database? Differentiate between SQL and NoSQL databases.

A database is an organized collection of structured information, or data, typically stored electronically in a computer system. A database is usually controlled by a database management system (DBMS).
<br>
Key Differences between SQL and NoSQL:-
<br>
SQL  is primarily called RDBMS or Relational Databases, whereas NoSQL is a Non-relational or Distributed Database.
<br>
SQL databases are table-based databases, whereas NoSQL databases can be document-based, key-value pairs, and graph databases.
<br>
SQL databases are vertically scalable, while NoSQL databases are horizontally scalable.
<br>
SQL databases have a predefined schema, whereas NoSQL databases use a dynamic schema for unstructured data.
<br>
SQL requires specialized DB hardware for better performance while NoSQL uses commodity hardware.

Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

DDL or Data Definition Language actually consists of the SQL commands that can be used to define the database schema. It simply deals with descriptions of the database schema and is used to create and modify the structure of database objects in the database. DDL is a set of SQL commands used to create, modify, and delete database structures but not data.
<br>
CREATE: This command is used to create the database or its objects (like table, index, function, views, store procedure, and triggers).
<BR>
eg: CREATE TABLE testdb.test_table(C1 int, C2 FLOAT, C3 VARCHAR(40));
<br>
DROP: This command is used to delete objects from the database. eg- DROP Table testdb.test_table;
    <br>

ALTER: This is used to alter the structure of the database. eg - ALTER TABLE testdb.test_table ADD C4 INT;
    <br>
TRUNCATE: This is used to remove all records from a table, including all spaces allocated for the records are removed. eg- TRUNCATE TABLE testdb.test_table;

Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.
<br>
The SQL commands that deals with the manipulation of data present in the database belong to DML or Data Manipulation Language.
<br>
INSERT : It is used to insert data into a table. eg- INSERT INTO testdb.test_table VALUES(1,74.5,"Aritra",2023);
<br>
UPDATE: It is used to update existing data within a table. eg- UPDATE testdb.test_table SET C4 = 2022 WHERE C1 = 1;
<br>
DELETE : It is used to delete records from a database table. eg- DELETE FROM testdb.test_table SET C4 = 2022 WHERE C1 = 1;

Q5. Explain Primary Key and Foreign Key.
<br>
Primary Key: A primary key is used to ensure that data in the specific column is unique. A column cannot have NULL values. It is either an existing table column or a column that is specifically generated by the database according to a defined sequence. 
<br>
A foreign key is a column or group of columns in a relational database table that provides a link between data in two tables. It is a column (or columns) that references a column (most often the primary key) of another table.
