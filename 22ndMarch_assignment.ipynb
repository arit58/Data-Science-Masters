Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result. <br>

To calculate the Pearson correlation coefficient between two variables, we need to compute the covariance and standard deviation of both variables.

Suppose we have a sample of n students, and for each student, we have recorded their study time x and their exam score y. Then, the Pearson correlation coefficient r is given by:

r = cov(x, y) / (std(x) * std(y))

where cov(x,y) is the covariance between x and y, and std(x) and std(y) are the standard deviations of x and y, respectively.

To interpret the result, the Pearson correlation coefficient r takes values between -1 and 1. A value of r = 1 indicates a perfect positive linear relationship between the two variables, meaning that as one variable increases, the other variable also increases in a linear fashion. A value of r = -1 indicates a perfect negative linear relationship, meaning that as one variable increases, the other variable decreases in a linear fashion. A value of r = 0 indicates no linear relationship between the two variables.

If r is between -1 and 0, or between 0 and 1, then there is a linear relationship between the two variables, but it is not perfect. The closer r is to -1 or 1, the stronger the linear relationship. The closer r is to 0, the weaker the linear relationship.

For example, if we calculate r to be 0.7, this indicates a strong positive linear relationship between study time and exam scores. In other words, students who spend more time studying tend to get higher exam scores, and the relationship between study time and exam scores is linear. On the other hand, if we calculate r to be -0.2, this indicates a weak negative linear relationship between study time and exam scores. In this case, there is a slight tendency for students who spend more time studying to get lower exam scores, but the relationship is not very strong or clear.
